```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

text = extract_text_from_pdf("fruit_info.pdf")
print(text[:1000])  # Preview first 1000 chars
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load sentence embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Split text into chunks
def split_text(text, chunk_size=500):
    return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

docs = split_text(text)
embeddings = model.encode(docs)
```

```python
import faiss

dimension = embeddings[0].shape[0]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

```python
from transformers import pipeline

qa_model = pipeline("text2text-generation", model="google/flan-t5-base")

def get_answer(question):
    question_embedding = model.encode([question])
    D, I = index.search(np.array(question_embedding), k=1)
    matched_text = docs[I[0][0]]

    prompt = f"Context: {matched_text}\n\nQuestion: {question}"
    result = qa_model(prompt, max_new_tokens=100)
    return result[0]['generated_text']
```

```python
print("🤖", get_answer("Which fruit is rich in potassium?"))
```

```python
import gradio as gr

demo = gr.Interface(
    fn=get_answer,
    inputs="text",
    outputs="text",
    title="📄 PDF Q&A (No LangChain!)",
    description="Ask questions from the uploaded PDF file."
)

demo.launch()
```
